# **Importing Libraries**

```python
import numpy as np
import pandas as pd
```

# Reading **CSV**

```python
df = pd.read_csv("flightprices.csv")
#df.head()
```

# **feature engg Date of Journey as Date, Month and Year**

```python
#feature engg Date of Journey as Date, Month and Year
df["Date"] = df["Date_of_Journey"].str.split('/').str[0]
df["Month"] = df["Date_of_Journey"].str.split('/').str[1]
df["Year"] = df["Date_of_Journey"].str.split('/').str[2]
df["Date"]= df["Date"].astype(int)
df["Month"]= df["Month"].astype(int)
df["Year"]= df["Year"].astype(int)
df.drop("Date_of_Journey",axis=1,inplace=True)
df.head(100)
```

```python
#thought to drop additional info but it would be required furhter
df["Additional_Info"].unique()
```

# **label encoded stop mapppings manually**

```python
#label encoded stop mapppings manually
stop_mappings = {
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
}
df["Total_Stops"] = df["Total_Stops"].map(stop_mappings).fillna(1).astype(int)
df.head()
```

# **feat engg departure time**

```python
#df['Dep_Time'].unique()
```

```python
#feature engg departure time
df["Dep hour"] = df["Dep_Time"].str.split(":").str[0]
df["Dep Min"] = df["Dep_Time"].str.split(":").str[1]
df["Dep hour"] = df["Dep hour"].astype(int)
df["Dep Min"]= df["Dep Min"].astype(int)
df.drop("Dep_Time", axis = 1, inplace = True)
df.head()
```

# **feat engg arrival time**

```python
#feature engg arrival time
df["Arrival hour"] = df["Arrival_Time"].str.split(':').str[0]
df["Arrival min"] = df["Arrival_Time"].str.split(':').str[1].str.split(' ').str[0]
df["Arrival hour"] = df["Arrival hour"].astype(int)
df["Arrival min"]= df["Arrival min"].astype(int)
month_mappings = {
    "Jan": 1,
    "Feb": 2,
    "Mar": 3,
    "Apr": 4,
    "May": 5,
    "Jun": 6,
    "Jul": 7,
    "Aug": 8,
    "Sep": 9,
    "Oct": 10,
    "Nov": 11,
    "Dec": 12,
    np.nan: 0
}
df["Arrival date"]= pd.to_numeric(df["Arrival_Time"].str.split(' ').str.get(1),errors = 'coerce').fillna(0).astype(int)
df["Arrival month"]= df["Arrival_Time"].str.split(' ').str.get(2).map(month_mappings).fillna(0).astype(int)
df.drop("Arrival_Time", axis = 1, inplace = True)
df.head()
```

```python
df['Duration'] = df['Duration'].str.replace('h', '*60').str.replace(' ', '+').str.replace('m', '*1').apply(eval)
```
